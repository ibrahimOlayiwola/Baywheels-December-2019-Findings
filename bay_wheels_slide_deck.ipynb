## Investigation Overview

>  The goal of this analysis is to get time most trips were taken and which of the users took it.

## Dataset Overview

> The dataset consist of 150102 rows and 14 columns.
- **duration_sec:** This is the trip duration in seconds.
- **start_time:** This column tells us the start time and date of the trip.
- **end_time:** This is the end time and date of the trip.
- **start_time_day:** The day of the week the trip was started.
- **end_station_id:** The identification number of the station the trip ended.

## Dataset Overview


- **start_station_name:** The name of the station the trip was started.
- **end_station_name:** The name of the station the trip ended.
- **start_station_latitude** and **start_station_longitude:** Address of the start station on a GPS.
- **start_date:**  The date of the month the trip was started/booked.
- **Bike_id:** Identification number of the bike.
- **User_type:** A user's status (Either a subcriber or a customer).
- **duration_min:** The duration of trip in minutes.**

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df_copy = pd.read_csv('df_master.csv')
df_copy.head(3)
```

## Distrubution of Trips for the Month

> - The visualization shows the amount of trips that was taken through each day in the month. It showed that the number of trips occurred from the at the beginning till around 20th of the month.


```python
base_color = sb.color_palette()[0]
plt.figure(figsize = [15, 5]);
g = sb.countplot(data=df_copy, x='start_date', color = base_color)
plt.title("Distrubution of Trips for the Month", y=1.05, fontsize=15, fontweight='medium');
plt.xlabel('Date of Month', y=1.05, fontsize=12, fontweight='medium');
plt.ylabel('Number of Trips', y=0.5, fontsize=12, fontweight='medium');
```

## Duration of Trips for Each Day

> The chart shows the duration on the days of the month. Surprisingly, the low number of trips towards the end of the month spent more time on the trips than the higher number of trips earlier in the month.

```python
plt.figure(figsize = [15, 5]);
ax = sb.lineplot(x="start_date", y="duration_min",sort=False, data=df_copy);
plt.title("Duration of Trips for Each Day", y=1.05, fontsize=15, fontweight='medium');
plt.xlabel('Date of Month', y=1.05, fontsize=12, fontweight='medium');
plt.ylabel('Duration of Trip', y=0.5, fontsize=12, fontweight='medium');
```

## Daily Duration of Trips by User Type

> The daily duration of trips by user types tells us which users, customers or subscribers spent more time on the trips on each day.

```python
plt.figure(figsize = [15, 5]);
ax = sb.lineplot(x='start_date', y='duration_min',
                  hue="user_type", data=df_copy);
plt.title("Daily Duration of Trips by User Type", y=1.05, fontsize=15, fontweight='medium');
plt.xlabel('Date of Month', y=1.05, fontsize=12, fontweight='medium');
plt.ylabel('Duartion of Trips', y=0.5, fontsize=12, fontweight='medium');
```

## Conclusion.
From the above charts. it can be said that the customer users use the most time on trips. It is also noted that the most number of trips were made during the first 20 days of the month but the most duration of trips was made during the last 10 days of the month.
